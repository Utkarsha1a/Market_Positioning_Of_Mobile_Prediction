```python
import seaborn as sns
columns=['battery_power', 'clock_speed', 'fc', 'int_memory', 'm_dep','mobile_wt', 'n_cores', 'pc', 'px_height', 'px_width', 'sc_h', 'sc_w','talk_time', 'price_range']
# draw the pairplot such that the diagonal should be density plot and the other graphs should be scatter plot
sns.pairplot(mob_data[columns], size=2, kind= "scatter", diag_kind="kde")

# display the plot
plt.show()
```

```python
# ploting correlation plot
corr = mob_data.corr()
# set the figure size
plt.figure(figsize=(30, 15))

# plotting the heat map
# corr: give the correlation matrix
# cmap: colour code used for plotting
# vmax: gives maximum range of values for the chart
# vmin: gives minimum range of values for the chart
# annot: prints the correlation values in the chart
# annot_kws: sets the font size of the annotation
# set condition to get a strong correlation between the variables
sns.heatmap(corr[(corr >= 0.6) | (corr <= -0.6)], 
            cmap='YlGnBu', vmax=1.0, vmin=-1.0,
            annot=True, annot_kws={"size": 15})

# set the title
# fontsize=30: set the font size of the title
plt.title('Correlation between features', fontsize=15)
# display the plot
plt.show()
```
