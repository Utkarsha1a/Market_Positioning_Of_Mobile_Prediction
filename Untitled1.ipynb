# <font color='red'>4 KNN</font>

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=10)
```

```python
k_range = range(1, 26)
scores = []
for n in k_range:
    knn=KNeighborsClassifier(n_neighbors=n)
    knn.fit(X_train, y_train)
    result = knn.score(X_test, y_test)
    y_pred = knn.predict(X_test)
    accuracy = metrics.accuracy_score(y_test, y_pred)
    scores.append(accuracy)
    #print('KNN score for n = {}  is {} and accuracy is {}'.format(n,round(result,2),round(accuracy,2)))
```

```python
# allow plots to appear within the notebook
%matplotlib inline
# plot the relationship between K and testing accuracy
# plt.plot(x_axis, y_axis)
plt.plot(k_range, scores,color='red',linestyle='dashed',marker='o',markerfacecolor='blue',markersize=10)
plt.xlabel('Value of K for KNN')
plt.ylabel('Testing Accuracy')
```

# <font color='red'>5 KNN with Feature Scaling</font>

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train_sc= sc.fit_transform(X_train)
X_test_sc= sc.transform(X_test)
```

```python
#X_train_sc
```

```python
scores2 = []
for n in k_range:
    knn_scale=KNeighborsClassifier(n_neighbors=n)
    knn_scale.fit(X_train_sc, y_train)
    result2 = knn_scale.score(X_test_sc, y_test)
    y_pred2 = knn_scale.predict(X_test_sc)
    accuracy2 = metrics.accuracy_score(y_test, y_pred2)
    scores2.append(accuracy2)
    #print('KNN score for n = {}  is {} and accuracy is {}'.format(n,round(result2,2),round(accuracy2,2)))
```

```python
# allow plots to appear within the notebook
%matplotlib inline
# plot the relationship between K and testing accuracy
# plt.plot(x_axis, y_axis)
plt.plot(k_range, scores2,color='red',linestyle='dashed',marker='o',markerfacecolor='blue',markersize=10)
plt.xlabel('Value of K for KNN with Feature Scaling')
plt.ylabel('Testing Accuracy')
```

# <font color='red'>6 KNN with GridSearch</font>

```python
knn = KNeighborsClassifier()

parameters = {'n_neighbors':[4,5,6,7],
              'leaf_size':[1,3,5],
              'algorithm':['auto', 'kd_tree'],
              'weights':['uniform','distance'],
              'metric':['euclidean'],
              'n_jobs':[-1]}

#Fit the modelX_train, X_test, y_train, y_test 
model = GridSearchCV(knn, param_grid=parameters)
model.fit(X_train,y_train)

#predictions on test data
y_predict3=model.predict(X_test)
```

```python
print (model.best_score_)
print (model.best_params_)
print (model.best_estimator_)
```

```python
accuracy2 = metrics.accuracy_score(y_test, y_predict3)
```

```python
accuracy2
```

```python
#List Hyperparameters that we want to tune.
leaf_size = list(range(1,50))
n_neighbors = list(range(1,30))
p=[1,2]

hyperparameters = dict(leaf_size=leaf_size, n_neighbors=n_neighbors, p=p)

knn_2 = KNeighborsClassifier()
clf = GridSearchCV(knn_2, hyperparameters, cv=10)
best_model = clf.fit(X_train,y_train)

print('Best leaf_size:', best_model.best_estimator_.get_params()['leaf_size'])
print('Best p:', best_model.best_estimator_.get_params()['p'])
print('Best n_neighbors:', best_model.best_estimator_.get_params()['n_neighbors'])
```

```python
#Select the most important features of the Mobile data set
#FEATURE SELECTION CODE:

from sklearn.ensemble import RandomForestClassifier
# Create a random forest classifier
clf = RandomForestClassifier(n_estimators=10000, random_state=0, n_jobs=-1)

# Train the classifier
clf.fit(X_train, y_train)


feat_labels = x.columns.values
# Print the name and gini importance of each feature
feature_importance = []
for feature in zip(feat_labels, clf.feature_importances_):
    #rint(feature)
    feature_importance.append(feature)

#feature_importance

from sklearn.feature_selection import SelectFromModel
# Create a selector object that will use the random forest classifier to identify
# features that have an importance of more than 0.01
sfm = SelectFromModel(clf, threshold=0.02)


# Train the selector
sfm.fit(X_train, y_train)


selected_features = []
# Print the names of the most important features
for feature_list_index in sfm.get_support(indices=True):
    selected_features.append(feat_labels[feature_list_index])


#selected_features
```
